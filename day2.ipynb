```
(

  df
  .groupby('param_marka-pojazdu')['price_value']
  .agg([np.mean, np.median, np.size])
  .sort_values(by='mean',ascending=False)
  .head(50)

).plot(kind='bar', figsize=(15,5), subplots=True)
```

```
 def group_and_barplot(group_feat, agg_feat='price_value', agg_funcs=[np.mean, np.median, np.size], sort_feat='mean', top=50, subplots=True):
  return (

    df
    .groupby(group_feat)[agg_feat]
    .agg(agg_funcs)
    .sort_values(by=sort_feat,ascending=False)
    .head(top)

  ).plot(kind='bar', figsize=(15,5), subplots=subplots)

group_and_barplot('param_marka-pojazdu');
```
